```python
file_path = '/home/xev/research/nfld/data/drug-list.xlsx'
data = pd.read_excel(file_path, engine='openpyxl', sheet_name='drug_list')
data
```

```python
from drugbank_scrape import scrape_drugbank
```

```python
data['Name'] = None
data['Desc'] = None
data['US Approved'] = None
data
```

```python
for idx, row in data.iterrows():
    drug = row['Drug']
    print(f"Scraping {drug}...")
    name, desc, us_approved = scrape_drugbank(drug)
    data.at[idx, 'Name'] = name
    data.at[idx, 'Desc'] = desc
    data.at[idx, 'US Approved'] = us_approved
```

```python
data
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a') as writer:
    data.to_excel(writer, sheet_name='drug_list_complete', index=False)
```
